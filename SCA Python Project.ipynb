# 1. write a program where the computer randomly generates a number          between 0 and 20.the user needs to guess what the number is. if the user guesses wrong, tell them their guess is either too hiigh or too low

```python
import random
number= random.randrange(0,20)
guess=int(input("Guess a number between 0 and 20: "))
while guess !=number:
    if guess<number:
        print("You need to guess higher. Try again please")
        guess=int(input("\nGuess a number between 0 and 20: "))
    else:
        print("You need to quess lower. Try again please")
        guess=int(input("\nGuess a number between 0 and 20: "))
print("You guess the number right!")
```

# 2. write a program which generates a random password for the user. ask that the password should be a minimum of 6 characters long.

```python
import random
uppercase_letters = ("ABCDEFGHIJKLMNOPRSTUVWXYZ")
lowercase_letters = uppercase_letters.lower()
digits = "123456789"
symbols = "!@#$%^&*()?/=+_-"

upper, lower, nums, syms = True, True, True, True

all = ""
if upper:
    all+= uppercase_letters
if lower:
    all += lowercase_letters
if nums:
    all += digits
if syms: 
    all += symbols
    
length = 6
amount = 12

for x in range(amount):
    password = "".join(random.sample(all, length))
    print(password)
```
